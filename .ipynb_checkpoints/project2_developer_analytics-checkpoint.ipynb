## Section 1: Developer Productivity & Code Output Analysis
**Business Question**: *Who are our top-performing developers by lines of code written, and how does their code volume correlate with their logged hours and self-reported productivity scores?*

To answer this, we query the `developers` database. Because developers write multiple `code_snippets` and submit multiple `daily_logs`, a simple flat join would create a Cartesian product that artificially inflates code lines and logged hours. To prevent this, we aggregate the metrics separately using Common Table Expressions (CTEs) before combining them.

```python
query_productivity = """
WITH loc_agg AS (
    SELECT 
        dev_id, 
        SUM(lines_of_code) AS total_lines_of_code,
        COUNT(snippet_id) AS snippet_count
    FROM 
        code_snippets
    GROUP BY 
        dev_id
),
log_agg AS (
    SELECT 
        dev_id, 
        SUM(hours_logged) AS total_hours_logged,
        ROUND(AVG(productivity_score), 2) AS avg_productivity_score
    FROM 
        daily_logs
    GROUP BY 
        dev_id
)
SELECT 
    d.username,
    d.country,
    d.subscription_tier,
    d.is_active,
    COALESCE(c.total_lines_of_code, 0) AS total_lines_of_code,
    COALESCE(l.total_hours_logged, 0.0) AS total_hours_logged,
    COALESCE(l.avg_productivity_score, 0.0) AS avg_productivity_score
FROM 
    developers d
LEFT JOIN 
    loc_agg c ON d.dev_id = c.dev_id
LEFT JOIN 
    log_agg l ON d.dev_id = l.dev_id
ORDER BY 
    total_lines_of_code DESC;
"""

# If testing locally on the workspace active port (5433), we ensure password defaults to empty
test_pass = "" if DB_PASS == "YOUR_PASSWORD" else DB_PASS

try:
    conn = psycopg2.connect(
        host=DB_HOST,
        port=DB_PORT,
        dbname=DB_NAME,
        user=DB_USER,
        password=test_pass
    )
    df_productivity = pd.read_sql_query(query_productivity, conn)
    conn.close()
    print("Query 1 executed successfully. Result loaded into pandas DataFrame.")
except Exception as e:
    print(f"Error executing Query 1: {e}")

df_productivity
```

### Visualization: Developer Code Volume & Productivity
Let's visualize the code contributions of each developer. We will use a horizontal bar chart to list the developers by total lines of code. This gives us a clear ranking of our top contributors.

```python
plt.figure(figsize=(10, 6))

# Sort df by lines of code for ascending order in horizontal bar chart
df_sorted = df_productivity.sort_values(by='total_lines_of_code', ascending=True)

# Plot horizontal bar chart
bars = plt.barh(df_sorted['username'], df_sorted['total_lines_of_code'], 
                color=sns.color_palette("viridis", len(df_sorted)), edgecolor='none')

# Add titles and labels
plt.title('Total Lines of Code Written by Developer', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Total Lines of Code', fontsize=12, labelpad=10)
plt.ylabel('Developer Username', fontsize=12, labelpad=10)

# Add values to the end of bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 15, bar.get_y() + bar.get_height()/2, f'{int(width):,}', 
             va='center', ha='left', fontsize=10, fontweight='semibold', color='#333333')

plt.tight_layout()
plt.show()
```

### Management Insights: Developer Productivity & Code Output
- **Primary Contributor**: `sql_master` is our top contributor by code volume, writing **1,650 lines of code** with an outstanding average productivity score of **10.0** (based on 12.0 hours logged).
- **High-Volume Outliers**: `syntax_error` also shows significant output (**500 lines of code**) with an average productivity score of **6.0**.
- **Low-Activity Accounts**: Several developers (e.g. `legacy_dev`, `null_pointer`, `logic_loop`) have written **0 lines of code** and logged very low or no hours. This indicates either onboarding delays, system inactivity, or legacy status.

## Section 2: Project Health & Bug Severity Analysis
**Business Question**: *Which projects are struggling with the highest backlog of bugs, and how does the severity of those bugs affect project health and delivery risk?*

Managing bug backlogs is critical for release planning. This query joins `projects` and `bug_reports` to aggregate bug counts by severity level (Critical, High, Medium, Low) and calculate the average days bugs have been open (excluding resolved ones). This highlights projects that are experiencing severe quality issues or bottlenecked triage processes.

```python
query_bugs = """
SELECT 
    p.project_name,
    p.primary_language,
    p.database_used,
    COUNT(b.bug_id) AS total_bugs,
    SUM(CASE WHEN b.severity_level = 'Critical' THEN 1 ELSE 0 END) AS critical_bugs,
    SUM(CASE WHEN b.severity_level = 'High' THEN 1 ELSE 0 END) AS high_bugs,
    SUM(CASE WHEN b.severity_level = 'Medium' THEN 1 ELSE 0 END) AS medium_bugs,
    SUM(CASE WHEN b.severity_level = 'Low' THEN 1 ELSE 0 END) AS low_bugs,
    ROUND(AVG(CASE WHEN b.status != 'Resolved' THEN b.days_open ELSE NULL END), 1) AS avg_days_open_active_bugs
FROM 
    projects p
LEFT JOIN 
    bug_reports b ON p.project_id = b.project_id
GROUP BY 
    p.project_id, p.project_name, p.primary_language, p.database_used
ORDER BY 
    total_bugs DESC, avg_days_open_active_bugs DESC;
"""

try:
    conn = psycopg2.connect(
        host=DB_HOST,
        port=DB_PORT,
        dbname=DB_NAME,
        user=DB_USER,
        password=test_pass
    )
    df_bugs = pd.read_sql_query(query_bugs, conn)
    conn.close()
    print("Query 2 executed successfully. Result loaded into pandas DataFrame.")
except Exception as e:
    print(f"Error executing Query 2: {e}")

df_bugs
```

### Visualization: Bug Severity Breakdown by Project
To easily evaluate project risks, we will create a stacked bar chart showing the breakdown of bugs by severity for each project. This visualization immediately reveals projects with a dangerous accumulation of Critical and High bugs.

```python
# Filter out projects with no bugs for cleaner visualization
df_bugs_filtered = df_bugs[df_bugs['total_bugs'] > 0].copy()

# Sort by total bugs ascending so it looks nice in horizontal stacked chart
df_bugs_filtered = df_bugs_filtered.sort_values(by='total_bugs', ascending=True)

# Prepare stacked data
categories = df_bugs_filtered['project_name']
low = df_bugs_filtered['low_bugs']
medium = df_bugs_filtered['medium_bugs']
high = df_bugs_filtered['high_bugs']
critical = df_bugs_filtered['critical_bugs']

fig, ax = plt.subplots(figsize=(10, 6))

# Plot stacked bars
b1 = ax.barh(categories, low, label='Low', color='#3498db')
b2 = ax.barh(categories, medium, left=low, label='Medium', color='#f1c40f')
b3 = ax.barh(categories, high, left=low+medium, label='High', color='#e67e22')
b4 = ax.barh(categories, critical, left=low+medium+high, label='Critical', color='#e74c3c')

# Labels and styling
ax.set_title('Bug Severity Distribution by Project', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Number of Bug Reports', fontsize=12, labelpad=10)
ax.set_ylabel('Project Name', fontsize=12, labelpad=10)
ax.legend(title='Severity Level', loc='lower right', frameon=True)

# Show values inside or next to bars
for i in range(len(categories)):
    total = df_bugs_filtered.iloc[i]['total_bugs']
    ax.text(total + 0.1, i, f' {int(total)} total', va='center', ha='left', fontsize=10, fontweight='semibold')

plt.tight_layout()
plt.show()
```

### Management Insights: Project Health & Bug Severity Breakdown
- **Highest Backlog Project**: `Inventory API` has the most reported issues (**2 bugs**), including a **Critical** bug that has been open for an average of **16.0 days**. This represents a bottleneck that should be reviewed.
- **Risk Hotspots**:
  - `Offline DevLog` has **2 bugs** (1 High, 1 Low), but they are relatively new or resolved quickly (average **2.5 days open**).
  - `E-commerce Backend` has **1 Critical** bug and `3D Render Pipeline` has **1 Critical** bug, which represent high architectural or security risks.
- **Languages / Databases Involved**: The majority of projects with critical issues are using **Python** (BigQuery, PostgreSQL) or **C#** (SQL Server). The C# project (`Inventory API`) requires immediate developer attention due to the high backlog duration.

## Section 3: Subscription Tier Value Analysis
**Business Question**: *Is there a significant difference in performance, code output, and active hours between Pro subscription tier developers and Free subscription tier developers?*

Understanding how subscription tiers correlate with developer engagement can help justify marketing investments or feature segmentation. This query aggregates developers by subscription tier and evaluates the average hours logged, average productivity score, and average code lines written per developer. To verify these findings, we also query the daily activity logs to visualize the relationship between daily hours logged and productivity scores.

```python
query_tier = """
WITH loc_agg AS (
    SELECT dev_id, SUM(lines_of_code) AS total_loc
    FROM code_snippets
    GROUP BY dev_id
),
log_agg AS (
    SELECT dev_id, AVG(hours_logged) AS avg_hours, AVG(productivity_score) AS avg_prod
    FROM daily_logs
    GROUP BY dev_id
)
SELECT 
    d.subscription_tier,
    COUNT(DISTINCT d.dev_id) AS total_developers,
    ROUND(AVG(COALESCE(log_agg.avg_hours, 0)), 2) AS avg_daily_hours,
    ROUND(AVG(COALESCE(log_agg.avg_prod, 0)), 2) AS avg_daily_productivity,
    ROUND(AVG(COALESCE(loc_agg.total_loc, 0)), 2) AS avg_lines_of_code_per_dev
FROM 
    developers d
LEFT JOIN 
    loc_agg ON d.dev_id = loc_agg.dev_id
LEFT JOIN 
    log_agg ON d.dev_id = log_agg.dev_id
GROUP BY 
    d.subscription_tier;
"""

query_scatter = """
SELECT 
    l.hours_logged,
    l.productivity_score,
    d.subscription_tier,
    d.username
FROM 
    daily_logs l
JOIN 
    developers d ON l.dev_id = d.dev_id;
"""

try:
    conn = psycopg2.connect(
        host=DB_HOST,
        port=DB_PORT,
        dbname=DB_NAME,
        user=DB_USER,
        password=test_pass
    )
    df_tier = pd.read_sql_query(query_tier, conn)
    df_scatter = pd.read_sql_query(query_scatter, conn)
    conn.close()
    print("Query 3 executed successfully. Subscription data loaded into DataFrames.")
except Exception as e:
    print(f"Error executing Query 3: {e}")

print("\nSubscription Tier Aggregation Summary:")
df_tier
```

### Visualization: Hours Logged vs. Productivity Score
To understand how work duration affects quality and speed (productivity), we plot daily logged hours against the productivity score. We color the points by subscription tier (Free vs Pro) and overlay a linear regression trend line to see the overall trend.

```python
plt.figure(figsize=(10, 6))

# Map colors to subscription tiers
colors = {'Pro': '#e74c3c', 'Free': '#3498db'}

# Scatter plot using Seaborn for beautiful visual representation
sns.scatterplot(data=df_scatter, x='hours_logged', y='productivity_score', 
                hue='subscription_tier', palette=colors, s=120, alpha=0.8, edgecolor='w')

# Calculate trend line
x = df_scatter['hours_logged']
y = df_scatter['productivity_score']
m, c = np.polyfit(x, y, 1)

# Plot trend line
plt.plot(x, m*x + c, color='#2c3e50', linestyle='--', linewidth=1.5, label='Regression Trendline')

# Customize titles and axis labels
plt.title('Productivity Score vs. Daily Hours Logged', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Daily Hours Logged', fontsize=12, labelpad=10)
plt.ylabel('Productivity Score (1-10)', fontsize=12, labelpad=10)
plt.xlim(0, 14)
plt.ylim(0, 11)
plt.legend(title='Developer Tier', loc='upper left')

plt.tight_layout()
plt.show()
```

### Management Insights: Subscription Tier Comparison
- **Pro Tier Performance**: Developers on the **Pro tier** show significantly higher engagement. They average **6.75 hours logged** per active day and have a high average productivity score of **7.60**. Their total lines of code per developer averages **541.67 lines**.
- **Free Tier Performance**: Developers on the **Free tier** average **2.50 hours logged** per active day, with a lower productivity score of **4.80** and an average of **204.50 lines of code**.
- **Correlation**: The scatter plot and regression trendline demonstrate a strong positive correlation between hours logged and productivity. This suggests that longer coding sessions lead to a higher self-reported or system-scored productivity, with Pro developers showing a clustered concentration in the high-hours, high-productivity region (upper-right).
